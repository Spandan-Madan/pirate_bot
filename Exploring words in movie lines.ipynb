```python
import json
from nltk.corpus import stopwords
from translate_module import translate_to_pirate as pirate_bot
```

```python
f = open('words_dictionary.json','rb')
words_dictionary = json.load(f)
f.close()
```

```python
f = open('data/movie_lines.txt','r')
content = f.readlines()
f.close()
```

```python
pirate_words_uncleaned = []
counter = 0
for c in content:
    c = c.rstrip()
    if counter % 500 == 0:
        print(c)
        translation = pirate_bot(c)
        print(translation)
        print('\n\n')
    counter +=1 
    words = c.split()
    filtered_words = [word for word in words if word not in stopwords.words('english')]
    for word in filtered_words:
        if word[-1].isalpha() == False:
            word = word[:-1]
        if len(word) > 2:
            if word.lower() not in words_dictionary.keys():
                pirate_words_uncleaned.append(word)
```

```python
from collections import Counter
c = Counter()
```

```python
c.update(pirate_words_uncleaned)
```

```python
c.most_common(1000)
```
